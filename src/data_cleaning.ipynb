```python
df = pd.read_csv('data/CMMD_clinicaldata_revision.csv')
df.head(60)
```

```python
df.info()
```

```python
# Strip leading and trailing spaces from all columns
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
# Loop over each column and print the unique values
for col in df.columns:
    print(col)
    print(df[col].unique())
```

```python
# Find duplicates in the 'ID1' column
duplicates = df[df.duplicated(subset=['ID1'], keep=False)]

# Count the frequency of each duplicated value in 'ID1'
count = duplicates['ID1'].value_counts()

# Print the duplicated values and their frequency
print("The following values in the 'ID1' column are repeated:")
for value, freq in count.items():
    print(f"{value}: {freq} times")
```

```python
# Find duplicates in the 'ID1' column
duplicates = df[df.duplicated(subset=['ID1'], keep=False)]

# Count the frequency of each duplicated value in 'ID1'
count = duplicates['ID1'].value_counts()

# Get the unique frequency counts
unique_counts = count.unique()

# Print the unique frequency counts
print('The unique frequency counts are:')
print(unique_counts)
```

```python
# Count the number of duplicate rows
num_duplicates = duplicates.shape[0]

# Print the number of duplicate rows
print(f"There are {num_duplicates} duplicate rows in the 'ID1' column.")
```

```python
# Get descriptive statistics for the 'Age' column
age_stats = df['Age'].describe()

# Print the statistics to the console
print('Descriptive statistics for the Age column:')
print(age_stats)
```

```python
# Concatenate the LeftRight column to the end of the ID1 values
df['ID1'] = df['ID1'] + df['LeftRight']
```

```python
# Convert values in the LeftRight column to 1 or 0
df['LeftRight'] = df['LeftRight'].apply(lambda x: 1 if x == 'R' else 0)
```

```python
# Rename the LeftRight column to Right
df = df.rename(columns={'LeftRight': 'right'})
```

```python
df.head()
```

```python
# Get the unique values in the 'Number' column
unique_values = df['number'].unique()

# Print the unique values to the console
print(unique_values)
```

```python
# Drop the 'Number' column
df = df.drop(['number'], axis=1)
```

```python
# Count the number of instances where 'classification' is 'Malignant' and 'ID1' starts with 'D1'
count = len(df[(df['classification'] == 'Malignant') & df['ID1'].str.startswith('D1')])

# Print the count to the console
print(count)

# Get the total number of ID1 values that start with 'D1'
count = df['ID1'].str.startswith('D1').sum()

# Print the count to the console
print(count)
```

```python
# Count the number of unique values in the ID1 column that start with 'D2'
count = df[df['ID1'].str.startswith('D2')]['ID1'].unique().size

# Print the count to the console
print(count)
```

```python
df.head()
```

```python
df = df[~df['ID1'].str.startswith('D2')]
df.info()
```

```python
# drop 'subtype' column
df.drop('subtype', axis=1, inplace=True)

# remove 'D1-' from beginning of values in 'ID1' column
df['ID1'] = df['ID1'].str.replace('D1-', '', regex=False).copy()

df.rename(columns={'ID1': 'ID', 'Age': 'age'}, inplace=True)

df.head(60)
```
